```python
import pandas as pd


data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank',
             'Grace', 'Helen', 'Ian', 'Jack', 'Karen', 'Leo'],

    'student_id': [101, 102, 103, 104, 105, 106,
                   107, 108, 109, 110, 111, 112],

    'major': ['CS', 'Math', 'CS', 'Physics', 'Math', 'CS',
              'Physics', 'Math', 'CS', 'Physics', 'Math', 'CS'],

    'attendance_score': [85.5, 78.0, 92.0, 88.5, 74.0, 90.0,
                         82.5, 79.5, 95.0, 87.0, 76.5, 91.0],

    'midterm_exam_score': [88, 75, 93, 85, 70, 91,
                           84, 77, 96, 83, 72, 89]
}

df = pd.DataFrame(data)

print("Dataset:n")
print(df)


grouped = df.groupby('major')['midterm_exam_score'].agg(['mean', 'median'])

print("nGrouped Statistics:n")
print(grouped)


highest_median_major = grouped['median'].idxmax()
highest_median_value = grouped['median'].max()

print("nMajor with highest median midterm score:")
print(f"{highest_median_major} (Median Score: {highest_median_value})")
```

The major with the highest median midterm exam score is CS, indicating that students in this group performed better on average compared to others.

```python
import pandas as pd

incomes = pd.Series([
    35000, 42000, 38000, 46000, 39000,
    41000, 37000, 45000, 43000, 40000,
    36000, 44000, 47000, 39000, 41000
])

print("Original Incomes:n")
print(incomes)


incomes_with_outlier = pd.concat([incomes, pd.Series([500000])], ignore_index=True)

print("nIncomes with Outlier:n")
print(incomes_with_outlier)


mean_income = incomes_with_outlier.mean()
median_income = incomes_with_outlier.median()

print("nMean Income:", mean_income)
print("Median Income:", median_income)
```

The mean is much higher than the median because the extreme outlier (500000) pulls the average upward. The median is less affected by extreme values, so it better represents the typical income.

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
population = np.random.normal(loc=70, scale=10, size=100000)


population = np.clip(population, 0, 100)


population_mean = np.mean(population)

print("Population Mean:", population_mean)



sample_means = []

for _ in range(200):
    sample = np.random.choice(population, size=30)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

sample_means = np.array(sample_means)



plt.figure(figsize=(8,5))
plt.hist(sample_means, bins=20, edgecolor='black')


plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')

plt.title("Histogram of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.legend()

plt.show()



mean_of_sample_means = np.mean(sample_means)

print("Mean of Sample Means:", mean_of_sample_means)
```

### Explanation

The mean of the sample means is very close to the population mean. This demonstrates the law of large numbers and shows that sampling variability decreases when we take many samples.
