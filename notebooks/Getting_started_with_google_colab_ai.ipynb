```python
import pandas as pd

# Step 1: Create dataset
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank',
             'Grace', 'Helen', 'Ian', 'Jack', 'Karen', 'Leo'],

    'student_id': [101, 102, 103, 104, 105, 106,
                   107, 108, 109, 110, 111, 112],

    'major': ['CS', 'Math', 'CS', 'Physics', 'Math', 'CS',
              'Physics', 'Math', 'CS', 'Physics', 'Math', 'CS'],

    'attendance_score': [85.5, 78.0, 92.0, 88.5, 74.0, 90.0,
                         82.5, 79.5, 95.0, 87.0, 76.5, 91.0],

    'midterm_exam_score': [88, 75, 93, 85, 70, 91,
                           84, 77, 96, 83, 72, 89]
}

df = pd.DataFrame(data)

# Step 2: Display dataset
print("Dataset:\n")
print(df)

# Step 3: Group by major and compute mean & median
grouped = df.groupby('major')['midterm_exam_score'].agg(['mean', 'median'])

print("\nGrouped Statistics:\n")
print(grouped)

# Step 4: Find major with highest median
highest_median_major = grouped['median'].idxmax()
highest_median_value = grouped['median'].max()

print("\nMajor with highest median midterm score:")
print(f"{highest_median_major} (Median Score: {highest_median_value})")
```
